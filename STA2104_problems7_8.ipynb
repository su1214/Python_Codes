```python
#1
world = pd.read_csv('world_small.csv',index_col="country", header=0, dtype={'gdppcap08':np.float64, 'polityIV':np.float64})
world.head()
```

```python
#6
world.loc[world.polityIV >= 20,'polityIV'] = np.nan
world.head(10)
```

```python
#problem8
#(1)
afr['population'] = np.random.randint(1000, 5000, len(afr))
print(afr.head())
```

```python
#(2)
world['gdp'] = world ['gdppcap08'].multiply( afr['population'])
world = world.fillna(0)
world.head()
```

```python
#(3)
world['gdp_log'] = np.log(world['gdppcap08'])
world.head()
```

```python
#(4)
world_num = world[['gdppcap08','polityIV']]
world_num = (world_num-world_num.mean())/world_num.std()
world_num.head()
```
