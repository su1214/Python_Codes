```python
import numpy as np

#1
np.random.seed(123)
aMat = np.random.randint(0, 100, (6, 10))
#(1)
aVec = np.where(np.mean(aMat, axis=0) > 50, 1, 0)
print(aVec)
#(2)
print(aMat[(aMat[:,4] > 50) | (aMat[:,6] < 30) , :])
#(3)
aMat[[3,1],[1,6]] = 999
print(aMat)
#(4)
print(np.mean(aMat[~np.any( aMat==999, axis=1),:],axis=0))
print(np.std(aMat[~np.any(aMat==999,axis=1),:],axis=0))
#(5)
print
aMat[aMat == 999] = np.mean(aMat, axis=0)[np.nonzero(aMat==999)[1]]
print(aMat)
#(6)
aMat2 = np.hstack([aMat, np.sum(aMat, axis=1).reshape(6,1)])
aMat2 = np.vstack([aMat2, np.sum(aMat2, axis=0)])
print(aMat2)
```

```python
#2
np.random.seed(123)
MAT = np.random.exponential(scale=1/5, size=(10, 7))
MAT= np.around( MAT, decimals=3 )
print(MAT)
#(1)
m= np.median(MAT)
Q1=np.percentile(MAT, q=25)
Q3=np.percentile(MAT, q=75)
print(MAT[(MAT<( Q1-1.5*(Q3-Q1))) | (MAT>( Q3+1.5*(Q3-Q1)))])
#(2)
mdiff = np.abs(MAT - m )
print(MAT[ mdiff== np.min(mdiff) ])
print(np.nonzero( mdiff== np.min(mdiff)))
#(3)
print(MAT[np.argsort(MAT[:,1])])
#(4)
print(MAT[np.argsort(np.mean(MAT,axis=0))[::-1]])
#(5)
print(MAT[[[0],[2]],[1,3,4]])
print(MAT[[0,0,0,2,2,2],[1,3,4,1,3,4]])
#(6)
MAT[(MAT<( Q1-1.5*(Q3-Q1))) | (MAT>( Q3+1.5*(Q3-Q1)))] = np.nan
print(MAT)
#(7)
print(MAT[np.sum( np.isnan(MAT), axis=1 ) > 0, : ])
```
